```python
from pychrtrace import gaussfit
import tifffile
import pandas as pd
import numpy as np
import glob
```

```python
folder = r'P:\ChromatinTeam\20210312_Franzi_Segmented_Spots_for_Kai\Spot_Volumes\\'
paths = glob.glob(folder+'*.tif')
names = [path.split('\\')[-1] for path in paths]
print(paths)
print(names)
```

```python
imgs = [tifffile.imread(path) for path in paths]
```

```python
fits = [gaussfit.fitSymmetricGaussian3D(img, 1, center=None)[0] for img in imgs]
```

```python
fits = pd.DataFrame(fits, columns=['BG', 'A', 'z','y','x','sigma_xy', 'sigma_z'])
fits['name'] = names
```

```python
fits.to_csv(folder+'results.csv')
```
